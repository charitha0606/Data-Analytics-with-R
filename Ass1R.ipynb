1.21 Load the data set primes (UsingR). This is the set of prime numbers in [1,2003]. How many are there? How many in the range [1,100]? [100,1000]?

```r
# Load the 'primes' dataset
data(primes)

# Check the number of prime numbers in [1, 2003]
total_primes <- length(primes)
cat("Total number of prime numbers in [1, 2003]:", total_primes, "\n")

# Check the number of prime numbers in [1, 100]
primes_1_to_100 <- primes[primes <= 100]
num_primes_1_to_100 <- length(primes_1_to_100)
cat("Number of prime numbers in [1, 100]:", num_primes_1_to_100, "\n")

# Check the number of prime numbers in [100, 1000]
primes_100_to_1000 <- primes[primes > 100 & primes <= 1000]
num_primes_100_to_1000 <- length(primes_100_to_1000)
cat("Number of prime numbers in [100, 1000]:", num_primes_100_to_1000, "\n")
```

```r
# Load the 'primes' dataset
data(primes)

# Find twin primes
twin_primes <- primes[which(diff(primes) == 2) + 1]

# Count the number of twin primes
num_twin_primes <- length(twin_primes)
cat("Number of twin primes:", num_twin_primes, "\n")
```

1.24 The data set mandms (UsingR) contains the targeted color distribution in a bag of M&Ms as percentages for varies types of packaging. Answer these questions.
1. Which packaging is missing one of the six colors?
2. Which types of packaging have an equal distribution of colors?
3. Which packaging has a single color that is more likely than all the others? What color is this?

```r
# Load the 'mandms' dataset
data(mandms)

# 1. Packaging Missing One of the Six Colors
# Check which packaging is missing one of the six colors (has a 0.0 value)
missing_color_packaging <- names(which(apply(mandms, 1, function(row) any(row == 0.0))))
cat("Packaging missing one of the six colors:", missing_color_packaging, "\n")

# 2. Packaging with Equal Color Distribution
equal_distribution_packaging <- rownames(mandms)[apply(mandms, 1, function(x) all(round(x, 2) == 16.67))]
cat("Packaging with equal color distribution:", equal_distribution_packaging, "\n")

# 3. Packaging with a Single Color More Likely Than Others
# Calculate the maximum percentage for each packaging
max_percentages <- apply(mandms, 1, max)

# Find the packaging with the highest maximum percentage
max_packaging <- rownames(mandms)[which.max(max_percentages)]

# Get the color with the highest percentage for the identified packaging
max_color <- colnames(mandms)[which.max(mandms[max_packaging, ])]

# Print the result
cat("Packaging with the highest likelihood for a single color:", max_packaging, "\n")
cat("Color with the highest percentage:", max_color, "\n")
```

1.25 The t imes variable in the data set nym. 2002 (UsingR) contains the time to finish for several participants in the 2002 New York City Marathon. Answer these questions.
1. How many times are stored in the data set?
2. What was the fastest time in minutes? Convert this into hours and minutes using R.
3. What was the slowest time in minutes? Convert this into hours and minutes using R.

```r
# Load the 'nym.2002' dataset
data(nym.2002)

# Extract the 'times' variable
times <- nym.2002$time

# Number of times stored
num_times <- length(times)
cat("Number of times stored:", num_times, "\n")

# Fastest time in minutes
fastest_time_minutes <- min(times)
cat("Fastest time in minutes:", fastest_time_minutes, "\n")

# Convert fastest time to hours and minutes
fastest_time_hours <- floor(fastest_time_minutes / 60)
fastest_time_remainder_minutes <- fastest_time_minutes %% 60
cat("Fastest time in hours and minutes:", fastest_time_hours, "hours", fastest_time_remainder_minutes, "minutes\n")

# Slowest time in minutes
slowest_time_minutes <- max(times)
cat("Slowest time in minutes:", slowest_time_minutes, "\n")

# Convert slowest time to hours and minutes
slowest_time_hours <- floor(slowest_time_minutes / 60)
slowest_time_remainder_minutes <- slowest_time_minutes %% 60
cat("Slowest time in hours and minutes:", slowest_time_hours, "hours", slowest_time_remainder_minutes, "minutes\n")
```

1.27 The data set uspop contains decade-by-decade population figures for the United States from 1790 to 1970.
1. Use names() and seq() to add the year names to the data vector.
2. Use diff() to find the inter-decade differences. Which decade had the greatest increase?
3. Explain why you could reasonably expect that the difference will always increase with each decade. Is this the case with the data?

```r
# Load the 'uspop' dataset
data("uspop")

# Add year names to the data vector
year_names <- seq(1790, 1970, by = 10)
names(uspop) <- year_names

# Find inter-decade differences
inter_decade_diffs <- diff(uspop)
greatest_increase_decade <- which.max(inter_decade_diffs)

# Output the decade with the greatest increase
cat("Decade with the greatest increase:", year_names[greatest_increase_decade + 1], "\n")

# Explanation and verification
# We expect the difference to increase with each decade due to population growth trends.
# Check if the differences indeed increase with each decade
is_increasing <- all(diff(inter_decade_diffs) > 0)
cat("Differences increase with each decade (as expected):", is_increasing, "\n")
```
